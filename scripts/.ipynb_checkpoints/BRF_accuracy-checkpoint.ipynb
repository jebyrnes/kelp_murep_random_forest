```python
file = 'kelp/BRF_accuracy.csv'

df = pd.read_csv(file)

# 2 to 4 m
x2 = df['Run type val'][df['Depth low (m)'] == 2][df['Number of Classes'] == 2]
y2 = df['Kelp Accuracy'][df['Depth low (m)'] == 2][df['Number of Classes'] == 2]

x3 = df['Run type val'][df['Depth low (m)'] == 2][df['Number of Classes'] == 3]
y3 = df['Kelp Accuracy'][df['Depth low (m)'] == 2][df['Number of Classes'] == 3]

# 4 to 6 m
x2_deep = df['Run type val'][df['Depth low (m)'] == 4][df['Number of Classes'] == 2]
y2_deep = df['Kelp Accuracy'][df['Depth low (m)'] == 4][df['Number of Classes'] == 2]

x3_deep = df['Run type val'][df['Depth low (m)'] == 4][df['Number of Classes'] == 3]
y3_deep = df['Kelp Accuracy'][df['Depth low (m)'] == 4][df['Number of Classes'] == 3]

#all depths
x2_all = df['Run type val'][np.isnan(df['Depth low (m)'])][df['Number of Classes'] == 2]
y2_all = df['Kelp Accuracy'][np.isnan(df['Depth low (m)'])][df['Number of Classes'] == 2]

x3_all = df['Run type val'][np.isnan(df['Depth low (m)'])][df['Number of Classes'] == 3]
y3_all = df['Kelp Accuracy'][np.isnan(df['Depth low (m)'])][df['Number of Classes'] == 3]

xlabel_string = ['R$_{rs}$','R$_{rs}$ + z', 'R$_{rs}$ + z \n+ slope \n+ rugosity']
                
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(12,4),sharey=True)
ax1.plot(x2,y2,marker = '.',color='k',label = 'Presence/Absence')
ax1.plot(x3,y3,marker = '.',color='k',linestyle=':',label = 'Low/Mod/Dense')
ax1.set_title('Depth = 2 - 4 m')
ax1.set_ylabel('Classification Accuracy',fontsize=14)
ax1.set_xlabel('Predictors',fontsize=14)
ax1.set_xticks([1,2,3])
ax1.set_xticklabels(xlabel_string) 
ax1.legend()

ax2.plot(x2_deep,y2_deep,marker = '.',color='k',label = 'Presence/Absence')
ax2.plot(x3_deep,y3_deep,marker = '.',color='k',linestyle=':',label = 'Low/Mod/Dense')
ax2.set_title('Depth = 4 - 6 m')
ax2.set_xlabel('Predictors',fontsize=14)
ax2.set_xticks([1,2,3])
ax2.set_xticklabels(xlabel_string)
ax2.legend(loc='lower right')

ax3.plot(x2_all,y2_all,marker = '.',color='k',label = 'Presence/Absence')
ax3.plot(x3_all,y3_all,marker = '.',color='k',linestyle=':',label = 'Low/Mod/Dense')
ax3.set_title('Depth = all')
ax3.set_xlabel('Predictors',fontsize=14)
ax3.set_xticks([1,2,3])
ax3.set_xticklabels(xlabel_string)
ax3.legend(loc='lower right')

plt.tight_layout()

fig.savefig('kelp/BRF_accuracy.jpg')
```
